```python
import pandas as pd


df = pd.read_csv("with_names_2.csv")
def get_chatres(x,x_type):
    all_obs= []
    for row in x:
        for line in row.split("\n"):
            if '\t' not in line:
                print('not formatted correctly')
            else:
                all_obs.append(line.strip().split("\t"))
    fandom_labeled = pd.DataFrame(all_obs)
    fandom_labeled.columns = ['name','fandom'+x_type]
    return fandom_labeled
fandom_labeled = pd.merge(get_chatres(df['chat_res.x'],'_claude'),
                             get_chatres(df['chat_res.y'],'_gemini'),
                             how='outer')

# Gemini was way better
fandom_labeled[(fandom_labeled.fandom_claude != fandom_labeled.fandom_gemini) &
                ((fandom_labeled.fandom_claude.str.lower() == 'ambiguous') | 
                 (fandom_labeled.fandom_gemini.str.lower() == 'ambiguous') )].sample(10)

fandom_labeled = get_chatres(df['chat_res.y'],'')
```
